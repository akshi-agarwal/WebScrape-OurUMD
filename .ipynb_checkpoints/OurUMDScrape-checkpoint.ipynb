```python
import requests
import urllib.request
from bs4 import BeautifulSoup
```

```python
url = 'http://www.ourumd.com/viewreviews/?all'
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
```

```python
tb = soup.findAll('table')
rows = tb[1].findAll('tr')
```

```python
for prof in rows[2:]:
    td = prof.findAll('td')
    print(td[2].text + "---" + td[1].text)
    
```
